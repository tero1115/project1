```python
"""아이리스 데이터셋csv파일 인터넷에서 다운로드했습니다.


공공데이터나 인터넷으로부터 csv파일을 다운로드하여 인공지능을 적용하기 위해 아래의 내용을 작성하시오.
다른 종류의 그래프를 1개 이상 생성(파이그래프 포함)하고 출력 결과들을 저장한 주피터노트북의 파일을 github에 'project1'이름으로 업로드
1) 데이터는 pandas형식으로 전환하기
2) Binning 수행하여 컬럼 추가하기
3) get_dummies()함수를 활용하여 컬럼 추가하기
4) NaN데이터가 있는지 확인하고 있다면 삭제 또는 min/max 알고리즘 적용하여 데이터프레임 업데이트하기
5) 인공지능 적용을 위해 모든 컬럼에 정규화(Normalization)을 적용
6) 결과를 3가지의 인공지능 모델에 적용하여 정확도(Accuracy)를 구하시오
7) 지금까지 준비한 결과를 기계학습의 입력으로 사용하고 지금까지 소개한 3가지 기계학습 유형을 선택하여 예측확률과 정답을 비교하시오."""
```

```python
#1) 데이터는 pandas형식으로 전환하기
import numpy as np
import pandas as pd
import networkx as nx

csv_data = pd.read_csv("dataset_Iris.csv",engine = 'python')
csv_data
```

```python
#2) Binning 수행하여 컬럼 추가하기
group_names = ['작다','보통','크다']
csv_data['PetalLengthGroup'],mybin = pd.cut(csv_data['PetalLengthCm'],3,labels=group_names,retbins=True)
csv_data
```

```python
#3) get_dummies()함수를 활용하여 컬럼 추가하기
csv_data = pd.concat([csv_data, pd.get_dummies(csv_data['PetalLengthGroup'])],axis=1,sort=False)
csv_data
```

```python
#4) NaN데이터가 있는지 확인하고 있다면 삭제 또는 평균을 구하여 대체하고 데이터프레임 업데이트하기

csv_data.isnull().sum() #각 컬럼별로 NaN갯수 확인
```

```python
#5) 인공지능 적용을 위해 모든 컬럼에 정규화(Normalization)을 적용[min-max]
csv_data['PetalLengthCm'] = (csv_data['PetalLengthCm']-csv_data['PetalLengthCm'].min())/(csv_data['PetalLengthCm'].max()- csv_data['PetalLengthCm'].min())
csv_data
```

```python
#6) 결과를 3가지의 인공지능 모델에 적용하여 정확도(Accuracy)를 구하시오
# 결정트리
iris_feature = csv_data[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']]
feature = np.array(iris_feature)
iris_names = csv_data['Species']
names = np.array(iris_names)
X = feature
y = names

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.5)

from sklearn import tree
clf = tree.DecisionTreeClassifier()
clf.fit(X_train, y_train)
predictions = clf.predict(X_test)
print('예측:\n', predictions)
print('정답(y_test):\n', y_test)

from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, predictions))
```

```python
# K 최근접 이웃
iris_feature = csv_data[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']]
feature = np.array(iris_feature)
iris_names = csv_data['Species']
names = np.array(iris_names)
X = feature
y = names

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.5)

from sklearn.neighbors import KNeighborsClassifier
clf = KNeighborsClassifier()
clf.fit(X_train, y_train)
predictions = clf.predict(X_test)
print('예측:\n', predictions)
print('정답(y_test):\n', y_test)

from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, predictions))
```

```python
# 서포트 벡터 머신
iris_feature = csv_data[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']]
feature = np.array(iris_feature)
iris_names = csv_data['Species']
names = np.array(iris_names)
X = feature
y = names

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.5)

from sklearn import svm
clf = svm.SVC(gamma=0.001, C=100.)
clf.fit(X_train, y_train)
predictions = clf.predict(X_test)
print('예측:\n', predictions)
print('정답(y_test):\n', y_test)

from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, predictions))
```
